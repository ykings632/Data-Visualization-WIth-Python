```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('San_2016.csv')
df.head()
```

```python
pd.DataFrame(df.groupby(['PdDistrict'])['IncidntNum'].count())
```

```python
df = pd.DataFrame(df.groupby(['PdDistrict'])['IncidntNum'].count()).reset_index()
df.head()
```

```python
df.rename(columns={'PdDistrict':'Neighborhood','IncidntNum':'Count'}, inplace=True)
```

```python
df.head()
```

```python
# Change datatype from Object to int
df['Count'] = df['Count'].astype(int)
```

```python
# sort the df
df.sort_values(by='Count', inplace=True, ascending=False)
```

```python
df
```

```python
# where reset index create a new index & by using drop it drops the previous index & holds the new one.
df.reset_index(drop=True, inplace=True)
```
